# 1 now lets import sentiment analysis using textblob method

```python
#!pip install textblob
```

```python
from textblob import TextBlob as tb
```

```python
sentiment=dataset["chat"].apply(lambda x:"positive" if tb(str(x)).sentiment.polarity>0
                                else "negative" if tb(str(x)).sentiment.polarity<0 
                                        else"neutral" )
score=dataset["chat"].apply(lambda x: tb(str(x)).sentiment.polarity)
#this is thee polaritz or the sentiment ofthat text
                                 

 # lambda dont support elif and supports only if else loop 
    #jus give lambda x : only one time
```

```python
# so wats happening above is instead of using for loop on all rows we r using .apply(lambda x:....)
# where .apply(..) applies to every row in that particular column
# lambda x: means we r asking it do sentimental analysis on each value in dataset[chat] where x is each value in col chat
# values r positve,negative,neutral
# and the analysis will happen only if we convert the string(x) to TextBlob
# .sentiment.polarity analyses the sentimenton x and works together after conversion to TextBlob
```

```python
dataset["sentiment"]=sentiment
dataset["scores"]=score
```

# 2 now lets do it using nltk method

```python
#!pip install nltk

# it is a natural language tool kit library
# used for tokenisation, sentiment analysis, word tagging etc
```

```python
import nltk
#nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# vader is an inbuilt sentiment analysis tool used in chats .it handles emojis,slangs etc
# and is fast and no prior traing is required.

# we r using vader to get compound scores,thefinal sentiment score.
```

```python
sia=SentimentIntensityAnalyzer()
```

```python
sia_scores=dataset["chat"].apply(lambda x: sia.polarity_scores(str(x)))

# here wats happening is polarity_scores from sia is beign found for every x using .apply(lambda x:...)
```

```python
sia_scores# we get a series.
```

```python
new=sia_scores.apply(pd.Series) # here wats happening we got a series of dictionary and wen u want to convert that to values with
#seperate columns then u do that_dict.apply(pd.series)
```

```python
new_data=pd.concat([dataset,new],axis=1)# this joins the table and the series new. 
# we can join 2 or more table or 2 or more series or tables and series.
```

```python
new_data# so now we get the compound value of each emotion also which will bee btw -1 to +1

# compound score is the normalised sum of score s which gives the final sentiment score 
#<.05 neg,>.05 pos btw neutral
```
