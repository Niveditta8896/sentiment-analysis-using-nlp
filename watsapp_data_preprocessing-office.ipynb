```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("office_group.txt",header=None,error_bad_lines=False,encoding="utf8")
```

```python
data
```

```python
data.columns
```

```python
dataset=data.drop([2, 3],axis=1)
```

```python
dataset.isna().sum()
```

```python
dataset[dataset[1]=="null"]
```

```python
dataset.fillna("",inplace=True)
```

```python
dataset
```

```python
dataset[0]=dataset[0].str.replace(r"http\S+", "", regex=True)
# this is a pattern r"http\S+" so watever u want to replace find a pattern of that structure and replace it.
```

```python
dataset
```

```python
dataset[0]=dataset[0].str.replace(r"\d{3}\s\d{8}","",regex=True)# this replaces regular expression like space s then digits
# this is a pattern for digits \d{3}\s\d{8} which we hv to replace
```

```python
dataset
```

```python
dataset=dataset.drop(0,axis=0)
```

# now we r going to split the table to date, time, messager, chat

```python
dataset.columns=["date","chat"]
```

```python
dataset.columns
```

```python
messagees=dataset["chat"].str.split("-",n=1,expand=True)# this is like sayin we r spliiting 1 time at "-"
messagees
```

```python
dataset["time"]=messagees[0]
dataset["chat"]=messagees[1]# then we add it to our original df by renaming
```

```python
dataset
```

```python
# now we r going to split chat to messager and chats
ch=dataset["chat"].str.split(":",n=1,expand=True)
ch
```

```python
dataset["messager"]=ch[0]
dataset["chat"]=ch[1]
```

```python
dataset
```

```python
dat=(dataset["chat"]=="None")| (dataset["messager"]=="None")|(dataset["chat"].isna()) | (dataset["messager"].isna())
#here we removed the rows with all values as null or none so we use or (|)
```

```python
new_data=dataset[~dat]
```

```python
new_data.isna().sum()
```

```python
new_data.to_csv("new_data_office.csv")
```
